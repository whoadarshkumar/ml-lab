```python
""" Clean the data and generate an insight on different parameters present in the dataset and give a clear picture on how the Instagram 
influencer can get more reach? Which parameter is more important for the influencer to work upon and why? Give a detailed explanation 
with proper statistical proof for Instagram_data.csv
 """



import pandas as pd

# Load the dataset
data = pd.read_csv("data.csv",encoding='cp1252')

data.dropna(inplace=True)
numeric_columns = ['Impressions', 'From Home', 'From Hashtags', 'From Explore', 'From Other', 'Saves', 
                    'Comments', 'Shares', 'Likes', 'Profile Visits', 'Follows']
data[numeric_columns] = data[numeric_columns].apply(lambda x: pd.to_numeric(x.replace(',', '')))
data.drop(axis=1,columns=['Caption','Hashtags'],inplace=True)
# # Calculate the sum of likes, comments, shares, and profile visits as Reach
data['Reach'] = data['Likes'] + data['Comments'] + data['Shares'] + data['Profile Visits']
# # Calculate correlation with the newly created Reach column
correlation = data.corr()

# print(correlation_with_reach)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(data)
plt.show()

correlation = data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Calculate correlation with reach
correlation_with_reach = correlation['Reach'].sort_values(ascending=False)
print(correlation_with_reach)
```

Here we can see that likes and impressions are very much related to reach so they are the most important attributes.
