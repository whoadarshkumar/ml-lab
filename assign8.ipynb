```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron, LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt
```

```python

# Load dataset
data = pd.read_csv('Data.csv')

# Split features and target variable
X = data.drop('M', axis=1)
y = data['M']
```

```python


# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Feature scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python


# Perceptron
perceptron = Perceptron()
perceptron.fit(X_train, y_train)
y_pred_perceptron = perceptron.predict(X_test)
accuracy_perceptron = accuracy_score(y_test, y_pred_perceptron)
report_perceptron = classification_report(y_test, y_pred_perceptron, zero_division=1)
```

```python


# SVM
svm = SVC()
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)
accuracy_svm = accuracy_score(y_test, y_pred_svm)
report_svm = classification_report(y_test, y_pred_svm, zero_division=1)

```

```python


# Logistic Regression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
y_pred_log_reg = log_reg.predict(X_test)
accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)
report_log_reg = classification_report(y_test, y_pred_log_reg, zero_division=1)
```

```python

# Plotting accuracy
models = ['Perceptron', 'SVM', 'Logistic Regression']
accuracies = [accuracy_perceptron, accuracy_svm, accuracy_log_reg]
plt.bar(models, accuracies, color=['blue', 'green', 'red'])
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Accuracy of Different Models')
plt.ylim(0, 1)
plt.show()
```

```python

# Printing classification reports
print("Perceptron Report:")
print(report_perceptron)
print("SVM Report:")
print(report_svm)
print("Logistic Regression Report:")
print(report_log_reg)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron, LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score

# Load dataset
data = pd.read_csv('wine.csv')

# Create binary labels based on 'quality' column
data['good/bad'] = data['quality'].apply(lambda x: 'good' if x > 6 else 'bad')
ct = (data['good/bad'] == 'good').sum()
print(ct)

# Split features and target variable
X = data.drop(['type', 'quality', 'good/bad'], axis=1)
y = data['good/bad']

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Feature scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Perceptron
perceptron = Perceptron()
perceptron.fit(X_train, y_train)
y_pred_perceptron = perceptron.predict(X_test)
accuracy_perceptron = accuracy_score(y_test, y_pred_perceptron)
report_perceptron = classification_report(y_test, y_pred_perceptron)

# SVM
svm = SVC()
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)
accuracy_svm = accuracy_score(y_test, y_pred_svm)
report_svm = classification_report(y_test, y_pred_svm)

# Logistic Regression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
y_pred_log_reg = log_reg.predict(X_test)
accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)
report_log_reg = classification_report(y_test, y_pred_log_reg)

# Plotting accuracy
models = ['Perceptron', 'SVM', 'Logistic Regression']
accuracies = [accuracy_perceptron, accuracy_svm, accuracy_log_reg]
plt.bar(models, accuracies, color=['blue', 'green', 'red'])
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Accuracy of Different Models')
plt.ylim(0, 1)
plt.show()

# Printing classification reports
print("Perceptron Report:")
print(report_perceptron)
print("SVM Report:")
print(report_svm)
print("Logistic Regression Report:")
print(report_log_reg)
```
