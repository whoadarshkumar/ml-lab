```python

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
iris_df = pd.read_csv('Iris.csv')

# a. Show size of the dataset
print("a. Size of the dataset:")
print(iris_df.shape)

# b. Show datatype for each column
print("\nb. Datatype for each column:")
print(iris_df.dtypes)

# c. Show distribution of data (use describe() function)
print("\nc. Distribution of data:")
print(iris_df.describe())

# d. Check if there are any null values
print("\nd. Check for null values:")
print(iris_df.isnull().sum())

# e. Check for duplicates
print("\ne. Check for duplicates:")
print(iris_df.duplicated().sum())


# f. Check the number of instances for each Species of flower
print("\nf. Number of instances for each Species:")
print(iris_df['Species'].value_counts())

# g. Compare sepal length and sepal width
print("\ng. Compare sepal length and sepal width:")
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=iris_df, hue='Species')
plt.show()

# h. Compare petal length and petal width
print("\nh. Compare petal length and petal width:")
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', data=iris_df, hue='Species')
plt.show()

# i. Use pairplot to show all comparisons
print("\ni. Pairplot for all comparisons:")
sns.pairplot(iris_df, hue='Species')
plt.show()

# j. Use histograms to compare sepal length, sepal width, petal length, and petal width across the Species

print("\nj. Histograms to compare features across Species:")
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.histplot(data=iris_df, x='SepalLengthCm', hue='Species', element='step', stat='density', common_norm=False)
plt.title('Sepal Length')

plt.subplot(2, 2, 2)
sns.histplot(data=iris_df, x='SepalWidthCm', hue='Species', element='step', stat='density', common_norm=False)
plt.title('Sepal Width')

plt.subplot(2, 2, 3)
sns.histplot(data=iris_df, x='PetalLengthCm', hue='Species', element='step', stat='density', common_norm=False)
plt.title('Petal Length')

plt.subplot(2, 2, 4)
sns.histplot(data=iris_df, x='PetalWidthCm', hue='Species', element='step', stat='density', common_norm=False)
plt.title('Petal Width')

plt.tight_layout()
plt.show()


# k. Use boxplot to show distribution of data across the Species
print("\nk. Boxplot to show distribution across Species:")
sns.boxplot(x='Species', y='SepalLengthCm', data=iris_df)
plt.show()

sns.boxplot(x='Species', y='SepalWidthCm', data=iris_df)
plt.show()

sns.boxplot(x='Species', y='PetalLengthCm', data=iris_df)
plt.show()

sns.boxplot(x='Species', y='PetalWidthCm', data=iris_df)
plt.show()

# l. Use violinplot to show distribution of data across the Species
print("\nl. Violinplot to show distribution across Species:")
sns.violinplot(x='Species', y='SepalLengthCm', data=iris_df)
plt.show()

sns.violinplot(x='Species', y='SepalWidthCm', data=iris_df)
plt.show()

sns.violinplot(x='Species', y='PetalLengthCm', data=iris_df)
plt.show()

sns.violinplot(x='Species', y='PetalWidthCm', data=iris_df)
plt.show()

```
